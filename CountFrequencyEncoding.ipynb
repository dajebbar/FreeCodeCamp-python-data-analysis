# Count Frequency Encoding Technique

Replace each of the categorical variable by the count

- Advantages: Curse of Dimensionality was not occured
- Disvantages: if two labels of categorical variable contain the same number of observations, they will be combined and may **loose valuable information**
  

```python
import numpy as np
import pandas as pd
```

```python
crop = pd.read_csv('./crop_production.csv')
crop.head()
```

```python
crop.select_dtypes(include='object').head()
```

```python
# transform each categorical serie to dictionary

def serie_to_dic(df, cat):
  return df[cat].value_counts().to_dict()
```

```python
serie_to_dic(crop, 'State_Name')
```
