## Apply the "Binary Encoder"

Binary Encoding converts a category into binary digit creates one columns. If there are **n** unique categories, then binary encoding results in the only **$log_2{n}$** features.

## Steps:
- The categories are first converted to numeric order starting from 1 (order is created as categories appear in a dataset and do not mean any ordinal natue).
- Then those integers are converted to binary digits code
- Then the digits of the binary number from separate column

```python
print(cat_features)
binary_cat = cat_features[:-2]
print(binary_cat)
```

```python
import category_encoders as ce

# create instance of binary encoding
encoder = ce.BinaryEncoder(df[binary_cat])

# fit and transform to get encoded cat
bin = encoder.fit_transform(df[binary_cat])
bin.head()
```

```python
df2 = pd.concat([bin, df.drop(columns=binary_cat)], axis=1)
df2.head()
```

```python
df2.columns
```

## Apply KNN Imputer

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)

df3 = df2.drop(columns=['Date']).copy()

df3[df3.columns] = imputer.fit_transform(df3[df3.columns])

df3.head()
```

```python
df3.isna().sum()
```

```python
df3 = pd.concat([df['Date'], df3], axis=1)
df3.head()
```
